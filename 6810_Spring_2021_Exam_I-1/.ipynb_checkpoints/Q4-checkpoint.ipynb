```python
res = []
for index, row in df.iterrows():
    for col in df.columns:
        if np.isnan(row[col]):
            continue
        res.append(index-(mu1+mu2*row[col]))
```

```python
eps = np.std(res)
eps
```

```python
alphas = [.1,.25,.5,.75,.9]
ts = [((15-mu1)/mu2)/(1-(norm.ppf(a)*sigma2/mu2)) for a in alphas]
ts
```

```python
B = 50000
n = 15
Fs = []
timepoints = np.arange(0,121000,120000/35)
for it in range(B):
    if it%100 == 1:
        print(it)
    df2 = pd.DataFrame(index=timepoints)
    mu2s = np.random.normal(mu2, sigma2, n)
    for i in range(n):
        df2[str(i+1)] = df2.index*mu2s[i]+mu1+np.random.normal(0,eps,36)
        
    datasets = []
    for col in df2.columns:
        temp = df2[col]
        temp2 = [[i] for i in temp.index]
        datasets.append((np.array(temp2),list(temp.tolist())))
    paras = []
    for ds in datasets:
        reg = LinearRegression().fit(ds[0], ds[1])
        paras.append((reg.intercept_,reg.coef_[0]))
    b0 = [i[0] for i in paras]
    b1 = [i[1] for i in paras]
    mu1_ = np.mean(b0)
    mu2_,sigma2_ = np.mean(b1), np.std(b1)
    Fs.append([(time-(15-mu1_)/mu2_)/(sigma2_*time/mu2_) for time in ts])
```

```python
F = [[norm.cdf(j) for j in i] for i in Fs]
```

```python
newF = sorted(F,key = lambda x:x[2])
```

```python
F_array = np.array(newF)
# B
alpha = 0.1
for i,(a,t) in enumerate(zip(alphas,ts)):
    curr_F = F_array[:,i]
    q = sum(curr_F<a)/B
    l = B*norm.cdf(2*norm.ppf(q)+2*norm.ppf(alpha/2))
    u = B*norm.cdf(2*norm.ppf(q)+2*norm.ppf(1-alpha/2))
#     print(q)
#     print(norm.ppf(q))
    print(t)
    print(a)
    print(curr_F[round(u)],curr_F[round(l)])
    print(l,u)
```
